```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Categories': ['YD MM IT', 'M3 IT', 'YD MM OOT', 'M3 OOT'],
    'Values': [11, 20.5, 15.9, 27.5]
}

# Create a DataFrame
met = pd.DataFrame(data)
```

```python


# Create the bar chart with custom colors
ax = met.plot(kind="bar", x='Categories', y='Values', color=['skyblue', 'salmon', 'skyblue', 'salmon'])

# Add labels to the axes
plt.xlabel('Model', fontsize=14)
plt.ylabel('MAPE', fontsize=14)
plt.figure(figsize=(8, 6))
```

```python
data = {
    'Categories': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 30]
}

# Create a DataFrame
met = pd.DataFrame(data)

# Create a Matplotlib figure and axis with custom figsize
fig, ax = plt.subplots(figsize=(8, 6))

# Create the bar chart with custom colors
met.plot(kind="bar", x='Categories', y='Values', color=['skyblue', 'salmon', 'lightgreen', 'lightcoral'], ax=ax)

# Add labels above the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black', xytext=(0, 10), textcoords='offset points')

# Add labels to the axes
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Set the title
plt.title('Bar Chart Example', fontsize=16)

# Show the plot
plt.show()
```
