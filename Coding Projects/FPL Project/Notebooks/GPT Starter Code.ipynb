```python
import requests

# Define the API endpoint URL
url = 'https://fantasy.premierleague.com/api/bootstrap-static/'

# Send a GET request to the API endpoint
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the JSON data from the response
    data = response.json()

    # Access the 'elements' key to get player information
    players = data['elements']

    # Create a dictionary to store player information
    player_info = {}

    # Iterate over each player and extract relevant information
    for player in players:
        player_id = player['id']
        player_info[player_id] = {
            'name': player['web_name'],
            'team': player['team'],
            'position': player['element_type'],
            'total_points': player['total_points'],
            'goals_scored': player['goals_scored'],
            # Add more attributes as needed
        }

    # Print the player information dictionary
    for player_id, info in player_info.items():
        print(f"Player ID: {player_id}")
        print(f"Player Name: {info['name']}")
        print(f"Team: {info['team']}")
        print(f"Position: {info['position']}")
        print(f"Total Points: {info['total_points']}")
        print(f"Goals Scored: {info['goals_scored']}")
        # Print more attributes as needed
        print("-------------------------------------")

else:
    print("Error occurred while accessing the API.")
```
